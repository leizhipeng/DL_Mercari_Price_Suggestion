```python
import numpy as np
import pandas as pd
import math
import matplotlib.pyplot as plt
%matplotlib inline 
```

```python
data = pd.read_table("train.tsv")
display(data.head(n=2))
print(data.shape)
```

```python
text = data.item_description.values
print(text[1])
```

```python
text_all = ' '.join(text)
```

```python
from wordcloud import WordCloud
# Generate a word cloud image
wordcloud = WordCloud().generate(text_all)

```

```python
# Display the generated image:
# the matplotlib way:
#import matplotlib.pyplot as plt
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")

```

```python
# lower max_font_size
wordcloud = WordCloud(max_font_size=40).generate(text_all)
plt.figure()
plt.imshow(wordcloud, interpolation="bilinear")
plt.axis("off")
#plt.show()
plt.savefig('word cloud.png', dpi=300, bbox_inches='tight')
```
